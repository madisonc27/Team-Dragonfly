First let's load up some basic files

```python
import pandas as pd
import numpy as np
```

Now let's load up our coffee data

```python
coffee_df = pd.read_csv('../data/arabica_data_cleaned.csv', index_col=[0])
```

Creating a copy of the original

```python
coffee = coffee_df.copy()
```

```python
coffee.head()
```

Now we remove the columns that we are not interested in.

```python
coffee.columns
```

```python
Unwanted = ['Region','Owner','Species', 'Farm.Name', 'Lot.Number', 'Mill',
       'ICO.Number', 'Company', 'Altitude', 'Producer',
       'Number.of.Bags', 'Bag.Weight', 'In.Country.Partner', 'Harvest.Year',
       'Grading.Date', 'Owner.1', 'Variety', 'Total.Cup.Points',
       'Moisture', 'Category.One.Defects', 'Quakers', 'Color',
       'Category.Two.Defects', 'Expiration', 'Certification.Body',
       'Certification.Address', 'Certification.Contact', 'unit_of_measurement']
```

```python
for categ in Unwanted:
    coffee.drop(categ,inplace=True, axis=1)
```

```python
coffee.head()
```

Renaming the columns to not include ".".

```python
coffee.columns = coffee.columns.str.replace(".", "_")
```

```python
coffee.head()
```

```python
coffee.info()
```

Note that many rows are missing altitude information. We can either perform imputation or remove those rows. We will try removing those rows, while also keeping a data frame that retains those rows for later imputation if we wish.

```python
coffee_dropnan = coffee.dropna(subset = ["altitude_low_meters", "altitude_high_meters", "altitude_mean_meters"], 
                                           axis = 0)
coffee_dropnan.info()
```

Now remove the countries with too few entries. This will be done in parallel for both data sets.

```python
countries_nan = coffee_dropnan.Country_of_Origin.value_counts()
countries = coffee.Country_of_Origin.value_counts()
```

```python
drop_list = []
drop_list_nan = []
```

```python
for count in countries.index:
    if countries[count] < 10:
        drop_list.append(count)
    
```

```python
for count in countries_nan.index:
    if countries[count] < 10:
        drop_list_nan.append(count)
```

```python
for item in drop_list:
    coffee.drop(coffee[coffee.Country_of_Origin == item].index, inplace=True)
```

```python
for item in drop_list_nan:
    coffee_dropnan.drop(coffee_dropnan[coffee_dropnan.Country_of_Origin == item].index, inplace=True)
```

```python
print(coffee.Country_of_Origin.value_counts())
print()
print(coffee_dropnan.Country_of_Origin.value_counts())
#notice that an additional country was dropped after removing the entries without altitude information
```

```python
drop_list
```

Removing any blank entries for Country of origin

```python
coffee['Country_of_Origin'].replace('',np.nan, inplace=True)
coffee_dropnan['Country_of_Origin'].replace('',np.nan, inplace=True)
```

```python
coffee.dropna(subset=['Country_of_Origin'], inplace=True)
coffee_dropnan.dropna(subset=['Country_of_Origin'], inplace=True)
```

Saving the files.

```python
coffee.to_csv('../data/coffee.csv',index = False)
coffee_dropnan.to_csv('../data/coffee_dropnan.csv',index = False)
```
