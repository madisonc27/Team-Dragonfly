# K Nearest Neighbors


```python
#first import packages we might need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

#### Load the data.

```python
from sklearn.model_selection import train_test_split
```

```python
## we'll start with the coffee data file
coffee_df = pd.read_csv("..\data\coffee.csv")
## check column names for the next step
coffee_df.columns
```

```python
## separate the columns into X for predictors and y for the country
predictors = ['Aroma', 'Flavor',
       'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean_Cup',
       'Sweetness', 'Cupper_Points']
X = coffee_df[predictors].copy()

y = coffee_df['Country_of_Origin']



```

```python
## perform the train test split
X_train, X_test, y_train, y_test = train_test_split(X.copy(), y.copy(),
                                                       shuffle=True,
                                                       random_state=214,
                                                       stratify=y, #be sure to stratify by the country
                                                       test_size=.2)
```

##### Create a pipeline for the model

```python
#import packages we will need
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
countries = coffee_df.Country_of_Origin.value_counts()
labels = countries.index
countries.count()
```

```python
kfold = StratifiedKFold(5, shuffle=True, random_state=14235)

knn_precs = np.zeros(5)

i = 0
for train_index, test_index in kfold.split(X_train, y_train.values):
    X_tt = X_train.iloc[train_index]
    X_ho = X_train.iloc[test_index]
    y_tt = y_train.iloc[train_index]
    y_ho = y_train.iloc[test_index]
    
    knn_pipe = Pipeline([('scale', StandardScaler()),
                    ('pca', PCA(10)), #this is a parameter that can be tuned   
                    ('knn', KNeighborsClassifier(n_neighbors = 7))]) #this is a parameter that can be tuned
    
    knn_pipe.fit(X_tt.values, y_tt.values)

    pred = knn_pipe.predict(X_ho.values)

    knn_precs[i] = precision_score(y_ho.values, pred, average = "macro") #the average can be changed as well
    conf_mat = confusion_matrix(y_ho.values, pred, labels=labels)


    i = i + 1
```

```python
pd.DataFrame(conf_mat,
                 columns = [ 'Predicted '+labels[0], 'Predicted '+labels[1], 'Predicted '+labels[2], 'Predicted '+labels[3],
                            'Predicted '+labels[4], 'Predicted '+labels[5], 'Predicted '+labels[6], 'Predicted '+labels[7], 
                            'Predicted '+labels[8], 'Predicted '+labels[9], 'Predicted '+labels[10], 'Predicted '+labels[11],
                            'Predicted '+labels[12], 'Predicted '+labels[13], 'Predicted '+labels[14], 'Predicted '+labels[15],
                            'Predicted '+labels[16], 'Predicted '+labels[17], 'Predicted '+labels[18]], 
                             index = ['Actual '+labels[0], 'Actual '+labels[1], 'Actual '+labels[2], 'Actual '+labels[3],
                            'Actual '+labels[4], 'Actual '+labels[5], 'Actual '+labels[6], 'Actual '+labels[7], 
                            'Predicted '+labels[8], 'Predicted '+labels[9], 'Predicted '+labels[10], 'Predicted '+labels[11],
                            'Predicted '+labels[12], 'Predicted '+labels[13], 'Predicted '+labels[14], 'Predicted '+labels[15],
                            'Predicted '+labels[16], 'Predicted '+labels[17], 'Predicted '+labels[18]])

```

```python
accuracy_score(y_ho, pred)
```

## Summary
Above we can see the confusion matrix for the predictions from the K nearest neighbors model. We see that the model is able to correctly predict 14 Mexico samples, 16 Colombia samples, and 11 Guatemala samples. These are the countries with the highest number of entries, so it makes sense they are more likely to be correctly predicted. Unfortunately, the algorithm only obtained an accuracy of 29% with the parameters we tested (these were not optimized), so we need to make some improvements.

## Trying again with a smaller data set
We also created a data set with a stricter cutoff for the required number of entries per country (50 instead of 10 used above). We will try the model again with this smaller data set to see if we can improve the accuracy.

```python
## import the new data set
coffee_df2 = pd.read_csv("..\data\coffee2.csv")
## create X and y again
## separate the columns into X for predictors and y for the country
X2 = coffee_df2[predictors].copy()

y2 = coffee_df2['Country_of_Origin']
```

```python
## perform the train test split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2.copy(), y2.copy(),
                                                       shuffle=True,
                                                       random_state=214,
                                                       stratify=y2, #be sure to stratify by the country
                                                       test_size=.2)
```

```python
countries2 = coffee_df2.Country_of_Origin.value_counts()
labels2 = countries2.index
countries2.count()
```

```python
kfold = StratifiedKFold(5, shuffle=True, random_state=14235)

knn_precs2 = np.zeros(5)

i = 0
for train_index, test_index in kfold.split(X_train2, y_train2.values):
    X_tt2 = X_train2.iloc[train_index]
    X_ho2 = X_train2.iloc[test_index]
    y_tt2 = y_train2.iloc[train_index]
    y_ho2 = y_train2.iloc[test_index]
    
    knn_pipe2 = Pipeline([('scale', StandardScaler()),
                    ('pca', PCA(10)), #this is a parameter that can be tuned   
                    ('knn', KNeighborsClassifier(n_neighbors = 7))]) #this is a parameter that can be tuned
    
    knn_pipe2.fit(X_tt2.values, y_tt2.values)

    pred2 = knn_pipe2.predict(X_ho2.values)

    knn_precs2[i] = precision_score(y_ho2.values, pred2, average = "macro") #the average can be changed as well
    conf_mat2 = confusion_matrix(y_ho2.values, pred2, labels=labels2)


    i = i + 1
```

```python
pd.DataFrame(conf_mat2,
                 columns = [ 'Predicted '+labels[0], 'Predicted '+labels[1], 'Predicted '+labels[2], 'Predicted '+labels[3],
                            'Predicted '+labels[4], 'Predicted '+labels[5], 'Predicted '+labels[6], 'Predicted '+labels[7]], 
                             index = ['Actual '+labels[0], 'Actual '+labels[1], 'Actual '+labels[2], 'Actual '+labels[3],
                            'Actual '+labels[4], 'Actual '+labels[5], 'Actual '+labels[6], 'Actual '+labels[7]])
```

```python
accuracy_score(y_ho2, pred2)
```

## Summary
We ran the pipeline again using the same parameters and here we obtained an accuracy of 38% compared to 29% obtained above. We can see that using a smaller data set with a stricter cutoff for country inclusion did increase the accuracy. However, this might be explained by the fact that Mexico, Colombia, and Guatemala, which are the countries with the largest number of samples, are now a larger proportion of the data. Looking at the confusion matrix above, we see that most countries are classifed as Mexico, Colombia, Guatemala, and Brazil. However, many of these classifications are incorrect. Unfortunately we don't seem to be able to use K Nearest Neighbors to predict the country of origin based on rating metrics such as aroma and balance.
